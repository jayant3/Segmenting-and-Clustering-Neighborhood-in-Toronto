```python
import requests
page = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
contents = page.content
```

```python
soup = BeautifulSoup(contents, 'html.parser')
```

```python
soup_str=soup.find_all(string=True)
```

```python
soup_str_1 = []
for i,n in enumerate(soup_str):
    if soup_str[i]!='\n':
        soup_str_1.append(soup_str[i])
```

```python
x=soup_str_1.index('M1A')
y=soup_str_1.index('M9Z')
print(x)
print(y)
```

```python
y=y+3
filtered_soup_str=soup_str_1[x:y]
```
